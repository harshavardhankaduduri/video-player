```python
df_new = df.drop(columns = ['name', 'host_name', 'last_review'])
```

```python
df_new.isna().sum()
```

```python
reviews_per_month_avg = df_new['reviews_per_month'].mean().round(2)
reviews_per_month_avg
```

```python
df_new.fillna(reviews_per_month_avg, inplace = True)
```

```python
df_new.isna().sum()
```

```python
df_new.duplicated().sum()
```

```python
df_new.head()
```

```python
text_columns = ['neighbourhood_group','neighbourhood', 'room_type']
for column in text_columns:
    df_new[column] = df_new[column].str.lower()
```

```python
df_new.head()
```

```python
from scipy.stats import zscore
```

```python
numerical_columns = ['price', 'minimum_nights',	'number_of_reviews', 'reviews_per_month', 'calculated_host_listings_count',	'availability_365']
z_scores = zscore(df_new[numerical_columns])
z_scores
```

```python
threshold = 3
```

```python
outliers = (z_scores > threshold).any(axis = 1)
df_new[outliers]
```

```python
outliers_index = df_new[outliers].index
df_cleaned = df_new.drop(index = outliers_index)
```

```python
print('Dataframe with outliers:', df_new.shape)
print('Dataframe without outliers:', df_cleaned.shape)
```

```python
df_cleaned
```

```python
df_cleaned.reset_index(drop = True, inplace = True)
```

The dataset was cleaned and preprocessed to ensure reliability for analysis. Irrelevant columns were removed, and missing values in  were filled with the column's mean. Text columns were standardized to lowercase, and outliers in numerical data were identified using Z-scores and removed. The resulting refined dataset is free of duplicates, missing values, and extreme outliers, making it ready for meaningful analysis.
